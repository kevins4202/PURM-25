```python
import pandas as pd
import dataframe_image as dfi
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python

# Load your CSV file
presence_df = pd.read_csv("tables/presence.csv")
stance_df = pd.read_csv("tables/stance.csv")
socialneeds_df = pd.read_csv("tables/social_needs.csv")
socialneeds_df
```

```python
presence_df
```

```python
plot_presence_df = presence_df[['macro_precision', 'macro_recall', 'macro_f1']]
plot_presence_df.columns = ["presence_"+col for col in plot_presence_df.columns]
plot_presence_df
```

```python
plot_stance_df = stance_df[['macro_precision', 'macro_recall', 'macro_f1']]
plot_stance_df.columns = ["stance_"+col for col in plot_stance_df.columns]
plot_stance_df
```

```python
plot_sn_df = socialneeds_df[['precision', 'recall', 'f1']]
plot_sn_df.columns = ["sn_"+col for col in plot_sn_df.columns]
plot_sn_df
```

```python
plot_all_df = pd.concat([socialneeds_df['model_prompt'], plot_presence_df, plot_stance_df, plot_sn_df], axis=1)
plot_all_df
```

```python
def highlight_max(s):
    is_max = s == s.max()
    return ['font-weight: bold' if v else '' for v in is_max]

# Style with bold max and center alignment
df_styled = plot_all_df.style.apply(highlight_max, axis=0).set_properties(**{
    'text-align': 'center',
    'border': '1px solid black',
    'padding': '5px',
    'font-size': '12pt'
}).set_table_styles([
    {'selector': 'thead th', 'props': [('background-color', '#f2f2f2')]},
    {'selector': 'tbody td', 'props': [('border', '1px solid #ccc')]}
])
df_styled
```
